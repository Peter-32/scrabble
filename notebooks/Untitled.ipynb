## Simulation for the Top 1000 Words to Know
- Randomize 7 letters (no bonus letters)
- Get the word score for each word
- Keep a running defaultdict total for each word
- Run simulations for a few hours


```python
all_answers = collections.defaultdict(int)

i = 0
while True:
    import random
    bag = ['A']*9+ ['B']*2+ ['C']*2+ ['D']*4+ ['E']*12+ ['F']*2+ ['G']*3+ ['H']*2+ ['I']*9+ ['J']*1+ ['K']*1+ ['L']*4+ ['M']*2+ ['N']*6+ ['O']*8+ ['P']*2+ ['Q']*1+ ['R']*6+ ['S']*4+ ['T']*6+ ['U']*4+ ['V']*2+ ['W']*2+ ['X']*1+ ['Y']*2+ ['Z']*1
    bag = [x.lower() for x in bag]
    # random.choice(bag)
    letter_set = random.sample(bag, 7)
    bonus_letters = []

    answers = []
    # i = 0    
    for word in words:
        if len(word) == 1:
            continue
    #     i = i + 1
    #     if i%100000 == 0:
    #         print(i)
        if letter_set_contains_word(letter_set, word, bonus_letters):
            answers.append((word, word_value(word)))
    
    # top 15
    answers = sorted(answers, key = lambda x: x[1], reverse=True)[0:15]
    for word, score in answers:
        all_answers[word] = all_answers[word] + score
    print(i)
    i += 1


```

```python
final_list = []
for key, value in all_answers.items():
    final_list.append((key, value))
final_list = sorted(final_list, key = lambda x: x[1], reverse=True)

# String
string = ""
for x,y in final_list:
    string = string + "{}:{}\n".format(x, y)

# Write
with open('../data/interim/best_vocab.txt', 'a') as file:
    file.write(string)
print(string)
```
